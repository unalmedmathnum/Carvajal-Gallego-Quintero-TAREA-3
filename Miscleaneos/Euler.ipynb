#### Implementemos método de Euler para sistemas de EDOs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Método de euler para unaa sola ecuación diferencial. Un poco por amor al arte
def funcion(t,y):
    return 1+t*np.sin(t*y)

def euler(func,ip, iv, fp, h):
    w = []
    t = []
    w.append(iv)
    p = ip
    y = iv
    while p < fp:
        z = h*func(p,y)
        y += z
        t.append(p)
        w.append(y)
        p += h
    t.append(p)
    return t,w

sol = euler(
    funcion,
    0,
    0,
    2,
    0.5
)

plt.figure(figsize=(12,5)) #Figura de solución
plt.subplot(121)
plt.plot(sol[0], sol[1], label='t')
plt.title('Ecuación (4): u(t)')
plt.grid(True)

plt.subplot(122) #Figura de derivada de solución
plt.plot(sol[0], sol[1], label='t')
plt.title("Ecuación (4): u'(t)")
plt.grid(True)
        
```

```python
#Euler para sistemas. Se corre en el ejemplo 1 de Burden. Se puede observar las diferencias con los resultados obtenidos en solve y RK, esto es mucho menos preciso
from euler_manual import eulerex

def ejemplo(t, y):
    return [
        y[1],
        np.exp(2 * t) * np.sin(t) - 2 * y[0] + 2 * y[1]
    ]


# Solve the system using Euler's method
sol = eulerex(
    ejemplo,
    0,
    [-0.4, -0.6],
    1,
    0.01
)

# Plot the results
plt.figure(figsize=(12, 5))
plt.subplot(121)
plt.plot(sol[0], sol[1], label='u(t)')
plt.title('u(t)')
plt.grid(True)

plt.subplot(122)
plt.plot(sol[0], sol[2], label="u'(t)")
plt.title("u'(t)")
plt.grid(True)

plt.show()
```
